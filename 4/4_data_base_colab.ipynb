### 1.3 実体-関連モデル (ER図)

情報の概念のモデル化に関する設計が重要
+ データ項目間の関係性をよく検討しておく必要がある

実体-関連モデル (ER)
実体は [四角], 属性は (楕円) で表現する．実体集合間の関連 は <ひし形> で表現する
例）[科目]:(科目名), (単位数)
    [学生]:(学生番号), (氏名)
    <履修>:[科目], [学生], (得点)

ERモデルから関係データベースへの変換ルール
+ 実体集合：属性をそのまま列にする
　関連：2つの実体の主キーと関連の属性を列にする
=> 変換後は，複数の実体集合をデータ項目としたviewとなる．
