## 1. ICT（情報通信技術）の進展，ビッグデータ

現在，スマートスピーカーや自動運転など，驚異的に発達した人工知能（Artifical Intelligence, AI）を活用した技術が私達の日常生活まで深く入りこんできています．
AIの導入は産業界においても積極的に進められており，小売業をはじめとして製造業にいたるまで，ありとあらゆる産業分野でビジネスの形態が変化しつつあります．
AIが適切に動作するためには，膨大な量のデータ，いわゆる「ビッグデータ」の存在が欠かせません [2]．

ここでは，AIにより引き起こされる社会や日常の大きな変化を認識します．
またAIが活用されている領域や現場，代表的な技術などを紹介します．
あわせて，AIに必須のデータがどのような形で収集され，いかに整理され分析されるのか，データ処理の大まかな流れについて学びます．

#### 2.2.1 研究開発におけるデータ・AI利活用 [2]

研究開発の領域では，技術開発や素材開発などにデータとAIが活用されている．

たとえば自動車開発においては，自動運転に向けた技術開発が行われている．
[自動車の位置情報や地図情報画像認識技術やセンシング技術（図2.2.2）](https://jidounten-lab.com/u_selfdriving-data-matome)を活用し，人の運転操作なしに走行できる自動車の開発を目指している．

素材開発においては，過去の実験データやシミュレーションデータから新素材を探索する取り組みが始まっている．
この新素材開発に関する取り組みは [マテリアルズインフォマティクス（図2.2.3）](https://jp.mitsuichemicals.com/jp/release/2021/2021_0628_03.htm) と呼ばれ，素材開発期間の短縮や未知の素材発見につながる可能性があると期待されている．

#### 2.2.2 購買物流・調達におけるデータ・AI利活用 [2]
