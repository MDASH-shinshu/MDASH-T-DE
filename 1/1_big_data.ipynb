## 1. ICT（情報通信技術）の進展，ビッグデータ

現在，スマートスピーカーや自動運転など，驚異的に発達した人工知能（Artifical Intelligence, AI）を活用した技術が私達の日常生活まで深く入りこんできています．
AIの導入は産業界においても積極的に進められており，小売業をはじめとして製造業にいたるまで，ありとあらゆる産業分野でビジネスの形態が変化しつつあります．
AIが適切に動作するためには，膨大な量のデータ，いわゆる「ビッグデータ」の存在が欠かせません [2]．

ここでは，AIにより引き起こされる社会や日常の大きな変化を認識します．
またAIが活用されている領域や現場，代表的な技術などを紹介します．
あわせて，AIに必須のデータがどのような形で収集され，いかに整理され分析されるのか，データ処理の大まかな流れについて学びます．
